```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python

# Use raw string literal to specify file path
file_path = r'C:\Users\vedan\OneDrive\Desktop\dataset and all\training_set_features.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Now you can work with the DataFrame 'df'
file_path2=r'C:\Users\vedan\OneDrive\Desktop\dataset and all\training_set_labels.csv'
df_labels=pd.read_csv(file_path2)
file_path3=r'C:\Users\vedan\OneDrive\Desktop\dataset and all\test_set_features.csv'
df_test=pd.read_csv(file_path3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
df.head()
```

```python
df_labels.head()
```

```python
x_train=df.drop(['respondent_id'],axis=1)
x_train.head()
```

```python
y_train=df_labels.drop(['respondent_id'],axis=1)
```

```python
X_test = df_test.drop(columns=['respondent_id'])
```

```python
categorical_features = x_train.select_dtypes(include=['object']).columns.tolist()
numerical_features = x_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
```

```python
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])
```

```python
model = MultiOutputClassifier(LogisticRegression(max_iter=1000))
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', model)])
pipeline.fit(x_train, y_train)
```

```python
# Predict probabilities
y_pred_prob = pipeline.predict_proba(X_test)
# y_pred_prob is a list of two arrays; one for each target variable
xyz_vaccine_prob = y_pred_prob[0][:, 1]  # Probability of xyz_vaccine == 1
seasonal_vaccine_prob = y_pred_prob[1][:, 1]  # Probability of seasonal_vaccine == 1

# Create a DataFrame with the predictions
predictions = pd.DataFrame({
    'h1n1_vaccine': xyz_vaccine_prob,
    'seasonal_vaccine': seasonal_vaccine_prob
},
index=df_test['respondent_id']
                           )
```
